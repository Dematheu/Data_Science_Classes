```python
# Importação das bibliotecas
import os
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregamento da base de dados
diretorio_atual = os.getcwd()
caminho_pasta_mae = os.path.abspath(os.path.join(diretorio_atual, ".."))
caminho_pasta_dados = os.path.join(caminho_pasta_mae, "dados")
caminho_arquivo = os.path.join(caminho_pasta_dados, "co2.csv")
base = pd.read_csv(caminho_arquivo)
base.head()
```

```python
# Gráfico de dispersão utilizando os atributos conc e uptake, agrupamento pelo type
srn.scatterplot(data=base, x='conc', y='uptake', hue='Type')
```

```python
# Seleção de registros específicos da base de dados (Quebec e Mississipi)
q = base.loc[base['Type'] == 'Quebec']
m = base.loc[base['Type'] == 'Mississippi']
```

```python
# Subgráfico (1 linha e duas colunas) mostrando gráficos sobre cada região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=q.conc, y=q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(x=m.conc, y=m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
# Refrigerado e não refrigerado
ch = base.loc[base['Treatment'] == 'chilled']
nc = base.loc[base['Treatment'] == 'nonchilled']
```

```python
# Gráfico somente com 'chilled' e 'nonchilled'
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(x=ch.conc, y=ch.uptake).set_title('Chilled')
plt.subplot(1,2,2)
srn.scatterplot(x=nc.conc, y=nc.uptake).set_title('Non chilled')
plt.tight_layout()
```
